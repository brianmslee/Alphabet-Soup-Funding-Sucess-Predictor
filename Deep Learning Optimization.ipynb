```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv('./Resources/charity_data.csv')
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
#  YOUR CODE GOES HERE
# 2nd try 
#application_df = application_df.drop(columns = ['EIN', 'NAME'])
#3rd try
application_df = application_df.drop(columns='EIN')
```

```python
#3rd try 
name_counts = application_df['NAME'].value_counts()
namelt5 = name_counts.loc[name_counts <= 5]
namelt5 = namelt5.index.to_list()
for x in namelt5:
    application_df['NAME'].replace(x,'Other', inplace=True)

application_df['NAME'].value_counts()
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE
#2nd try
#application_df.nunique()
#app_counts = application_df['APPLICATION_TYPE'].value_counts()
#applt500 = app_counts.loc[app_counts < 500].index.to_list()
#applt500
#3rd try
app_counts = application_df['APPLICATION_TYPE'].value_counts()
applt500 = app_counts.loc[app_counts < 500].index.to_list()
applt500
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#YOUR CODE GOES HERE
#2nd try
#application_types_to_replace = applt500
#application_types_to_replace

#Replace in dataframe
#for app in application_types_to_replace:
    #application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

#Check to make sure binning was successful
#application_df['APPLICATION_TYPE'].value_counts()
#3rd try 
application_types_to_replace = applt500
application_types_to_replace

#Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

#Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE
class_counts = application_df['CLASSIFICATION'].value_counts()
print(class_counts.min(), class_counts.median(), class_counts.max(), class_counts.mean())
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE
#2nd try
#classlt500 = class_counts.loc[class_counts < 500].index.to_list()
#3rd try
classlt500 = class_counts.loc[class_counts <= 1000].index.to_list()
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE
#2nd try
#classifications_to_replace = classlt500

# Replace in dataframe
#for cls in classifications_to_replace:
    #application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
#application_df['CLASSIFICATION'].value_counts()
#3rd try 
classifications_to_replace = classlt500

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
#2nd try 
#application_df.head()
#application_with_dummies_df = pd.get_dummies(application_df)
#3rd try 
application_df.head()
application_with_dummies_df = pd.get_dummies(application_df)
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE
#2nd try
#y = application_with_dummies_df["IS_SUCCESSFUL"].values
#X = application_with_dummies_df.drop(["IS_SUCCESSFUL"], axis= 1).values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
#X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 42)
#3rd try 
X = application_with_dummies_df.drop(["IS_SUCCESSFUL"], axis= 1).values
y = application_with_dummies_df["IS_SUCCESSFUL"].values

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state= 78)
```

```python
#2nd try 
# Create a StandardScaler instances
#scaler = StandardScaler()

# Fit the StandardScaler
#X_scaler = scaler.fit(X_train)

# Scale the data
#X_train_scaled = X_scaler.transform(X_train)
#X_test_scaled = X_scaler.transform(X_test)

#3rd try 
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
#2nd try 
#nn = tf.keras.models.Sequential()
#number_of_input_features = len(X_train[0])
#hidden_node1 = 100
#hidden_node2 = 50

# First hidden layer
#  YOUR CODE GOES HERE
#nn.add(tf.keras.layers.Dense(units = hidden_node1, activation = 'relu', input_dim = number_of_input_features))

# Second hidden layer
#  YOUR CODE GOES HERE
#nn.add(tf.keras.layers.Dense(units = hidden_node2, activation = 'sigmoid'))
#nn.add(tf.keras.layers.Dense(units = 30, activation = 'sigmoid'))

# Output layer
#  YOUR CODE GOES HERE
#nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
#nn.summary()

#3rd try 
nn = tf.keras.models.Sequential()
number_of_input_features = len(X_train[0])
hidden_node1 = 100
hidden_node2 = 50

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units = hidden_node1, activation = 'relu', input_dim = number_of_input_features))

# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units = hidden_node2, activation = 'sigmoid'))
nn.add(tf.keras.layers.Dense(units = 30, activation = 'sigmoid'))

# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
#2nd try
#nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
#3rd try 
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
#2nd try
#model = nn.fit(X_train_scaled, y_train, epochs = 100)
#3rd try 
model = nn.fit(X_train_scaled, y_train, epochs = 100)
```

```python
#2nd try
#model_loss, model_accuracy = nn.evaluate(X_test_scaled ,y_test,verbose=2)
#print(model_loss, model_accuracy)
#3rd try
model_loss, model_accuracy = nn.evaluate(X_test_scaled ,y_test,verbose=2)
print(model_loss, model_accuracy)
```
